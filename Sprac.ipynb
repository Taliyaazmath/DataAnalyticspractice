```python
import numpy as np
import pandas as pd
from scipy.spatial.distance import euclidean
```

```python

dataset = pd.read_csv('Mall_Customers.csv')

dataset.head()



```

```python
def euclidean_distance(row1_data, row2_data):
  return np.sqrt(np.sum((row1_data - row2_data)**2))

row_index1 = 1
row_index2 = 184

row1_data = dataset.iloc[row_index1, 4:].values
row2_data = dataset.iloc[row_index2, 4:].values

print(row1_data)
print(row2_data)
distance = euclidean_distance(row1_data, row2_data)

print(f"Euclidean Distance between Row {row_index1} and Row {row_index2}: {distance}")

def convert_to_similarity(distance):
  return 1 / (1 + distance)

similarity = convert_to_similarity(distance)

print(f"Similarity between Row {row_index1} and Row {row_index2}: {similarity}")
```
