```python
import matplotlib.pyplot as plt
import csv
Names = []
Values = []
with open('/blooprs.csv','r') as csvfile:
  lines = csv.reader(csvfile, delimiter=',')
  for row in lines:
    Names.append(row[0])
    Values.append(int(row[1]))
plt.scatter(Names, Values, color = 'g',s = 100)
plt.xticks(rotation = 25)
plt.xlabel('Names')
plt.ylabel('Values')
plt.title('Patients Blood Pressure Report', fontsize = 20)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
for dirname, _, filenames in os.walk('/content/data'):
  for filename in filenames:
    print(os.path.join(dirname, filename))
def catigorize_col (df, col, labels):
  edges = [df[col].describe()['min'],
          df[col].describe()['25%'],
          df[col].describe()['50%'],
          df[col].describe()['75%'],
          df[col].describe()['max']]
  df[col] = pd.cut(df[col], edges, labels = labels, duplicates='drop')
  return df
df_red = pd.read_csv('/winequality-red.csv')
df_red.head()
```

```python
df_white = pd.read_csv('/winequality-white.csv')
df_white.head()
```

```python
color_red = np.repeat('red', df_red.shape[0])

color_white = np.repeat('white', df_white.shape[0])
df_red['color'] = color_red
df_red.head()
```

```python
df_white['color'] = color_white
df_white.head()
```

```python
wine_df = pd.concat([df_white, df_red], axis = 0, ignore_index = True)
wine_df.head()
```

```python
wine_df['color'] = wine_df['color'].astype('category')
wine_df['color'].dtype
```
